#Writing a function to create responses using a prompt

```python
def create_text(text):
  response = openai.Completion.create(
      engine="text-davinci-002",
      prompt =  text,
      temperature = 0.6,
      top_p = 1,
      max_tokens = 64,
      frequency_penalty = 0,
      presence_penalty = 0
      )
  return print(response.choices[0].text)
```

Testing the API

```python
print(create_text("what is GPT-3"))
```
